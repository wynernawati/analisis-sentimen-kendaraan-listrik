```python
!pip install snscrape
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time

query = "kendaraan listrik OR motor listrik OR mobil listrik lang:id until:2023-02-28 since:2022-09-01 -filter:links"
tweets = []
limit = 4500

%time
try:
    print("Start Crawling")
    for tweet in sntwitter.TwitterSearchScraper(query=query).get_items():
        if len(tweets)==limit:
            break
        else:
            tweets.append([tweet.date, tweet.user.username, tweet.content])
    df = pd.DataFrame(tweets, columns=['datetime','username','content'])
except Exception as e:
    print(e)
print("finished")
```
